In this notebook I will be aggregating a years worth of listings, reviews, and calendar information from [Airbnb](http://insideairbnb.com/get-the-data.html). This data pertains to the San Francisco(SF) area from November of 2018 through October of 2019.

The Airbnb data descriptions are as follows:
- Listings: Detailed listings data including full descriptions and average review score
- Calendar: Detailed calendar data for listings, including listing id and the price and availability for that day;
- Reviews: Detailed review data for listings including unique id for each reviewer and detailed comments;

```python
#Read in libraries
import pandas as pd
import glob

#Set path to location of SF Airbnb raw data on local machine
path = r'C:\Users\kishe\Documents\Data Science\Projects\Python Projects\In Progress\Air BnB - SF\Data\01_Raw\SF Airbnb Raw Data/'
```

**SF Listings Data Aggregation(11/2018 - 10/2019)**

```python
#Capture listings data from path
all_files = glob.glob(path + "listings?*.gz")

#Create empty list to append csv files
li = []

#Read in listings data and append to li
for filename in all_files:
    df = pd.read_csv(filename, index_col=None, header=0)
    li.append(df)
```

```python
#Create aggregated dataframe
listings = pd.concat(li, sort=True, axis=0, ignore_index=True)

#Drop any duplicate rows
listings.drop_duplicates(inplace=True)

#View listings shape
print('Listings shape: ', listings.shape)

#Check to make sure columns have been properly concatenated
listings.head()
```

```python
#Set path to write df to csv
path_to_file = r'C:\Users\kishe\Documents\Data Science\Projects\Python Projects\In Progress\Air BnB - SF\SF Airbnb Raw Data - Aggregated\12_26_2019_Listings.csv'

#Write to csv file
listings.to_csv(path_to_file, sep=',')
```

**SF Reviews Data Aggregation(11/2018 - 10/2019)**

```python
#Capture reviews data
all_files = glob.glob(path + "reviews?*.gz")

#Create empty list
li = []

#Read in reviews data and append to li
for filename in all_files:
    df = pd.read_csv(filename, index_col=None, header=0)
    li.append(df)

#Create aggregated dataframe
reviews = pd.concat(li, sort=True, axis=0, ignore_index=True)

#Drop duplicate rows
reviews.drop_duplicates(inplace=True)

#View Shape
print('Reviews shape: ', reviews.shape)

#Check to make sure columns have been properly concatenated
reviews.head()
```

```python
#Set path to write df to csv
path_to_file = r'C:\Users\kishe\Documents\Data Science\Projects\Python Projects\In Progress\Air BnB - SF\SF Airbnb Raw Data - Aggregated\12_26_2019_Reviews.csv'

#Write to csv file
reviews.to_csv(path_to_file, sep=',')
```

**SF Calendar Data Aggregation(11/2018 - 10/2019)**

```python
#Capture calendar data
all_files = glob.glob(path + "calendar?*.gz")

#Create empty list
li = []

#Append reviews files to aggregate into a dataframe
for filename in all_files:
    df = pd.read_csv(filename, index_col=None, header=0)
    li.append(df)

#Create data frame
calendar = pd.concat(li, sort=True, axis=0, ignore_index=True)

#Drop any duplicate rows
calendar.drop_duplicates(inplace=True)

#View Shape
print('Calendar shape: ', calendar.shape)

#Check to make sure columns have been properly concatenate
calendar.head()
```

```python
#Set path to write df to csv
path_to_file = r'C:\Users\kishe\Documents\Data Science\Projects\Python Projects\In Progress\Air BnB - SF\SF Airbnb Raw Data - Aggregated\12_26_2019_Calendar.csv'

#Write to csv file
calendar.to_csv(path_to_file, sep=',')
```
